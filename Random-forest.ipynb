# Get the data

```python
import os
import tarfile
from six.moves import urllib

HOUSING_PATH = os.path.join("datasets", "housing")
```

```python
import pandas as pd
#print(pd.__version__) #print pandas version, e.g., 0.24.2
#to upgrade: pip install --upgrade pandas

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "./MergeConflictsDataset.csv")
    return pd.read_csv(csv_path,header = 0, delimiter=';')
```

```python
from sklearn.preprocessing import LabelEncoder

labelEnc = LabelEncoder()
data = housing.apply(labelEnc.fit_transform)
data
```

```python
data1 = data[housing.columns.drop('conflict')]
data1
```

```python
data['conflict']
```

```python
import numpy as np

np.random.seed(22)
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data, test_size=0.2, random_state=22)
```

```python
train_set.head()
```

```python
train_set1 = train_set[housing.columns.drop('conflict')]
train_set1.head()
```

```python
train_set['conflict']
```

```python
test_set.head()
```

```python
test_set1 = test_set[housing.columns.drop('conflict')]
test_set1.head()
```

```python
test_set['conflict']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree

scaler = StandardScaler()
train_set1 = scaler.fit_transform(train_set1)
test_set1 = scaler.transform(test_set1)

random = RandomForestClassifier(n_estimators=100, random_state=22)
random1 = RandomForestClassifier(n_estimators=150, random_state=22)
random2 = RandomForestClassifier(n_estimators=50, random_state=22)
random.fit(train_set1, train_set['conflict'])
random1.fit(train_set1, train_set['conflict'])
random2.fit(train_set1, train_set['conflict'])

dtc = tree.DecisionTreeClassifier(criterion='entropy', max_depth=18)
dtc.fit(train_set1,train_set['conflict']) 
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(random.estimators_[0], filled=True)
plt.savefig('random_forest.pdf')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
pred = random.predict(test_set1) 
pred1 = random1.predict(test_set1)
pred2 = random2.predict(test_set1)
pred_dtc = dtc.predict(test_set1)
```

```python
accuracy_score(test_set['conflict'], pred)
```

```python
recall_score(test_set['conflict'], pred)
```

```python
precision_score(test_set['conflict'], pred)
```

```python
val1 = recall_score(test_set['conflict'], pred)
val2 = recall_score(test_set['conflict'], pred1)
val3 = recall_score(test_set['conflict'], pred2)
val4 = recall_score(test_set['conflict'], pred_dtc)

my_dict = { 'random100': val1, 'random150': val2, 'random50': val3, 'decision_tree': val4}
sortedDict = sorted(my_dict)
print(sortedDict) #valorile recall_score sortate
```
